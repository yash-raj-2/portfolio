```python
# chage the date col data type from str to datetime

df.opening_date_clean = pd.to_datetime(df['opening_date_clean'])
```

```python
# rename col for better readability 

df = df.rename(columns = { 'coaster_name' : 'Coaster_Name',
                      'year_introduced' : 'Year_Introduced',
                      'latitude' : 'Latitude',
                      'longitude' : 'Longitude',
                      'opening_date_clean' : 'Opening_Date',
                      'speed_mph' : 'speed_mph', 
                      'height_ft' : 'Height_ft',
                      'Inversions_clean' : 'Inversions',
                      'Gforce_clean' : 'Gforce'
                    })
```

```python
# check if data has null values

df.isnull().sum()
```

```python
# check for duplicated values

df.duplicated()
df[df.duplicated()]
```

```python
# checking duplcate values in a col 

df[df.duplicated(subset=['Coaster_Name'])]
```

```python
df.loc[df['Coaster_Name'] =='Crystal Beach Cyclone']
```

```python
# finding dupliactes in subset of col 

df.loc[df.duplicated(subset=['Coaster_Name','Location','Opening_Date'])]
```

```python
# doing the inverse using tilde "~" to get not duplicated (unique ) values

df.loc[~df.duplicated(subset=['Coaster_Name','Location','Opening_Date'])]
```

```python
# resettng the index after removing the duplicated values

df = df.loc[~df.duplicated(subset=['Coaster_Name','Location','Opening_Date'])].reset_index(drop=True)
df.shape
```

## Answer a question aboput the data 
### What are the locations with the fastest rollercoaster 

```python
df.head()
```

```python
df['Location'].value_counts()
new_df = df.query('Location != "Other"')
```

```python
new_df = new_df.groupby('Location')['speed_mph'].agg(['mean', 'count'])
```

```python
new_df = new_df.query('count>=10').sort_values(by='mean' )
new_df
```

```python
ax = new_df['mean'].plot(kind='barh')
ax.set_xlabel('Avg Speed')
```
